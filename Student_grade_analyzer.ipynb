In this program we made a student grade analyzer using python and numpy , we take data from multiple students , store the marks in numpy and calculate 3 things which are total marks , avg marks and highest and lowest marks too . after this we assign grade based on the avg of the marks .


Logic: we first import NumPy into the code , then we find avg from the provided numbers , then we need to provide grades according to the avg , so we use if , elif and else function.We form an empty list to store marks and then an array , we continue this and then we go into a loop to find the avg . Once the avg is found the grade calculated will be according to the if ,elif and else function done above then we print all the things we need to calculate according to the question .


```python

import numpy as np


def find_grade(avg):

    if avg >= 85:
        return "A"

    elif avg >= 70:
        return "B"

    elif avg >= 50:
        return "C"

    else:
        return "Fail"



n = int(input("Enter number of students: "))


marks_list = []


for i in range(n):

    mark = float(input("Enter marks of student " + str(i + 1) + ": "))
    marks_list.append(mark)


marks = np.array(marks_list)


total_marks = np.sum(marks)

average_marks = np.mean(marks)

highest_marks = np.max(marks)

lowest_marks = np.min(marks)


grade = find_grade(average_marks)


print("\n----- STUDENT MARKS ANALYSIS -----")

print("Marks Entered:", marks)

print("Total Marks:", total_marks)

print("Average Marks:", average_marks)

print("Highest Marks:", highest_marks)

print("Lowest Marks:", lowest_marks)

print("Grade:", grade)
```
